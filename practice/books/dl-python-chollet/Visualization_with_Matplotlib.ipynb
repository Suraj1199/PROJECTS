# **Subplots**

* ### MATLAB-Style Interface

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)   # row , column, axes
plt.plot(x, np.sin(x), 'k', label='$\sin({x})$')
plt.legend(frameon=True)

plt.subplot(2, 2, 2)   
plt.plot(x, np.cos(x), 'r', label='$\cos({x})$')
plt.legend(frameon=True)

plt.subplot(2, 2, 3)   
plt.plot(x, np.sqrt(x), 'b', label='$\sqrt{x}$')
plt.legend(frameon=True)

plt.subplot(2, 2, 4)   
plt.plot(x, np.exp(x), 'g', label='$e^{x}$')
plt.legend(frameon=True)
```

`plt` can be used for small, straight forward plots. We use `ax` for more control over custom plots

* ### Object-Oriented Interface

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

ax[0, 0].plot(x, np.sin(x), 'k', label='$\sin({x})$')
ax[0, 0].legend(frameon=True)

ax[0, 1].plot(x, np.cos(x), 'r', label='$\cos({x})$')
ax[0, 1].legend(frameon=True)

ax[1, 0].plot(x, np.sqrt(x), 'b', label='$\sqrt{x}$')
ax[1, 0].legend(frameon=True)

ax[1, 1].plot(x, np.exp(x), 'g', label='$e^{x}$')
ax[1, 1].legend(frameon=True)
```

**More advanced customisation**

```python
fig = plt.figure()
ax = plt.axes(xlim=[0, 5], ylim=[0, 5]) # standard axes

ax1 = fig.add_axes([0.2, 0.2, 0.25, 0.25], sharex=ax, sharey=ax)
ax2 = fig.add_axes([0.6, 0.6, 0.25, 0.25], xlim=[0,10], ylim=[0, 10])
ax3 = fig.add_axes([0.2, 0.6, 0.25, 0.25], xlim=[5,10], sharey=ax2)
ax4 = fig.add_axes([0.6, 0.2, 0.25, 0.25], sharex=ax3, ylim=[-5, 5],)
```

# **Visualising Errors**

```python
# DATA
x = np.linspace(0, 10, 15)
dy = 0.8
dx = 0.4

y = np.sin(x) + dy * np.random.rand(15)
```

* ### **Basic Errorbars**

```python
plt.errorbar(x, y, xerr=dx, yerr=dy, fmt='ok')
```

**Cleaner error plot**

```python
plt.errorbar(x, y, 
             xerr=dx,
             yerr=dy, 
             fmt='o', 
             color='gray', 
             ecolor='lightgray', 
             elinewidth=3, 
             capsize=5, 
             ms=10, 
             mec='k', 
             mew=2)
```

* ### **Continuous Error**

    **combining `plt.plot` with `plt.fill_between`**

```python
xfit = x[:45]
yfit = np.sin(xfit) * xfit
dyfit = 0.5 * (np.hstack([2 * yfit[:15], 1.1 * yfit[15:30], 1.2 * yfit[30:]]) + np.random.rand(45) * 0.75)
```

```python
plt.figure(figsize=(10, 4))

plt.plot(xfit, yfit, c='k', lw='2')
plt.plot([3.15, 6.3], [0.01, 0.01], 'or', ms=10, mec='k')
plt.fill_between(xfit, yfit + dyfit, yfit - dyfit, color='lightgray')

plt.xlim([0, 8.5])
```

# **Density and Contour Plots**

```python
# DATA
xmin, xmax, ymin, ymax = 0, 5, 0, 5
x = np.linspace(xmin, xmax, 100)
y = np.linspace(ymin, ymax, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
```

* ### **`plt.contour`**

```python
plt.contour(X, Y, Z, colors='k')
plt.colorbar()
```

**NOTE:** When single color is used, positive values (peaks) are represented by solid lines whereas negative values (valleys) are represented by dashed lines. We can color-code the lines by specifying a colormap.

```python
plt.contour(X, Y, Z, cmap='RdGy')
plt.colorbar()
```

**Draw more lines in between**

```python
plt.contour(X, Y, Z, levels=15, cmap='RdGy')
plt.colorbar()
```

**NOTE:** The spaces between lines are quite distracting. So we mostly use filled contour plots
    <br> <br>

* ### **`plt.contourf`**

```python
plt.contourf(X, Y, Z, levels=15, cmap='RdGy')
plt.colorbar()
```

**NOTE:** The colour steps are still a bit discrete rather than continuous. A better way is to use `plt.imshow`, which interprets 2D grid of data as an image.
<br> <br>

* ### **`plt.imshow`**

```python
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy')
plt.colorbar()
```

**NOTE:** 
* No `X` and `Y` grid. Manually need to specify the extent (`[xmin, xmax, ymin, ymax]`) of the image.
* Image is inverted horizontally since default origin (0, 0) is in upper left for a standard image. Manually need to change origin to lower left

```python
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy', origin='lower')
plt.colorbar()
```

<br> **Combine Contour plots and Image plots**

```python
plt.figure(figsize=(7, 7))

contours = plt.contour(X, Y, Z, levels=3, colors='k')   # store the contour lines returned by the plot
plt.clabel(contours, fontsize=8, inline=True)   # customise contour laber
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy', origin='lower', alpha=0.5)

plt.colorbar(label='heights')
```

## **Two-Dimensional Histogram**

```python
mean = [0, 0] 
cov = [[1, 1], [1, 2]]

x, y = np.random.multivariate_normal(mean, cov, 1000).T
```

* ### **`plt.hist2d`**

```python
_ = plt.hist2d(x, y, bins=15, cmap='Blues') # hiding the output array and only displaying the plot
plt.colorbar(label='counts in bins')
```

* ### **`plt.hexbin`**

```python
plt.hexbin(x, y, bins=20, gridsize=15, cmap='Blues')  # gridsize analogous to bins
plt.colorbar(label='counts in bins')
```

* ### **Kernel Density Estimation (KDE)**

```python
from scipy.stats import gaussian_kde
```

```python
data = np.vstack([x, y])
kde = gaussian_kde(data)
```

```python
xgrid = np.linspace(-3.5, 3.5, 60)
ygrid = np.linspace(-6, 6, 60)

Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Zgrid = kde.evaluate([Xgrid.ravel(), Ygrid.ravel()]).reshape(Xgrid.shape)
```

```python
plt.imshow(Zgrid, extent=[-3.5, 3.5, -6, 6], origin='lower', cmap='Blues', aspect='auto')
plt.colorbar(label='Density')
```

# **Text and Annotation**

```python
plt.style.use('seaborn-whitegrid')
```

```python
# DATA 
x = np.arange(0, 100)
y = np.random.randn(100)
y = np.hstack([0.4 * y[:30], 0.2 * y[30:50], 2 * y[50:55], 0.1 * y[55:80], 0.3 * y[80:]])

min_idx = np.argmin(y)
max_idx = np.argmax(y)
```

* ### Using **`ax.text`**

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y,)
ax.axis([0, 100, -5, 5])

ax.text(min_idx - 1, y[min_idx], 'minimum', ha='right', color='k', size=15) ####
ax.text(max_idx + 1, y[max_idx], 'maximum', ha='left' , color='k', size=15) ####
```

* ### Using **`ax.annotate`**

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y)
ax.axis([0, 100, -5, 5])

ax.annotate('minimum', xy = (min_idx, y[min_idx]), ha='right', color='k', size=15)
ax.annotate('maximum', xy = (max_idx, y[max_idx]), xytext = (max_idx-3, y[max_idx]+0.25), color='k', size=15)
```

<hr>
