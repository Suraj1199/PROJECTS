````python
import matplotlib.pyplot as plt
plt.style.use('classic')

import numpy as np
import pandas as pd
````

````python
x = np.linspace(0, 10, 100)
y = np.cumsum(np.random.randn(100, 6), 0)
````

````python
plt.plot(x, y)
plt.legend(labels='ABCDEF', ncol=2, loc='best')
````

**Overwrite Matplotlib's default paramaters with Seaborn**

````python
import seaborn as sns
sns.set() 
````

````python
plt.plot(x, y)
plt.legend(labels='ABCDEF', ncol=2, loc='best')
````

# Relplot

````python
sns.relplot(x = x, y = np.sin(x)) # default = scatter
````

````python
sns.relplot(x = x, y = np.sin(x), kind='line')
````

# Histogram

````python
data = np.random.multivariate_normal([0, 2], [[4, 2], [2, 2]], 500)
data = pd.DataFrame(data, columns=['x', 'y'])
data.head()
````

````python
sns.histplot(data=data, alpha=0.5)
````

# KDE Plot

````python
sns.kdeplot(data=data, shade=True)
````

### 2-D Kde

````python
sns.kdeplot(data=data, x = 'x', y = 'y', shade=True)
````

Combine histogram with kde by passing keyword argument ```ked=True``` in `sns.histplot(...)`

````python
sns.histplot(data, kde=True, alpha=0.5)
````

# Violin Plot

````python
sns.violinplot(data=data, palette=['lightblue', 'lightgreen'])
````

# Joint Plot
joint distribution and marginal distribution

````python
sns.jointplot(data=data, x = 'x', y = 'y', color='c')
````

* ### kind kde

````python
sns.jointplot(data = data, x = 'x', y = 'y', kind='kde', color='r')
````

* ### kind hex

````python
with sns.axes_style('white'):
    sns.jointplot(data=data, x = 'x', y = 'y', kind='hex', color='b')
````

* ### kind reg

````python
sns.jointplot(data = data, x = 'x', y = 'y', kind='reg')
````
