Importing Matplotlib and setting Styles

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('seaborn-white')

%matplotlib inline
```

Importing numpy and pandas

```python
import numpy as np
import pandas as pd

x = np.linspace(0, 20, 100)
```

## Simple Plots

* ### MATLAB-Style Interface

```python
plt.plot(x, np.zeros(x.shape))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

* ### Object-Oriented Interface

```python
ax = plt.axes()

ax.plot(x, np.zeros(x.shape))
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### Line colors and Styles

```python
line_colors = ['blue',         # color name
               'g',            # color code
               '0.75',         # grayscale value between 0 and 1
               '#FFDD34',      # Hex code (RRGGBB from 00 to FF)
               (1, 0.2, 0.4),  # RGB tuple each value from 0 to 1
               'chartreuse'    # HTML color names 
              ]
```

```python
# code for better visualisation
plt.figure(figsize=(10, 5))
plt.axis([0, 20, -1.5, 1.5])

for i, line_color in enumerate(line_colors):
    plt.plot(x, np.sin(x - i), color=line_color)
```

```python
line_styles = ['solid', '-', 
               'dashed', '--', 
               'dashdot', '-.', 
               'dotted', ':']
```

```python
plt.figure(figsize=(10, 5))

for i, style in zip([0, 1, 5, 6, 10, 11, 15, 16], line_styles):
    plt.plot(x, x + i, linestyle=style)
```

#### String combination of line styles and colors

```python
combinations = ['-r',  # solid red'
                '--g', # dashed green
                '-.k', # dashdot black
                ':c',  # dotted cyan
               ]
```

```python
plt.figure(figsize=(10, 5))

for i, str_comb in enumerate(combinations):
    plt.plot(x, x + i * 2, str_comb)
```

### Labeling Plots

* #### MATLAB-Style Interface with `plt`

```python
plt.plot(x, np.sin(x))

plt.xlabel('x')
plt.ylabel('sin(x)')

plt.xticks(np.linspace(0, 20, 5))
plt.yticks(np.linspace(-1, 1, 5))

plt.title('A Sine Curve')
```

* #### Object-Oriented Interface with `ax`

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))

ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

ax.set_xticks(np.linspace(0, 20, 5))
ax.set_yticks(np.linspace(-1, 1, 5))

ax.set_title('A Sine Curve')
```

## Simple Scatter Plots with `plt.plot`

```python
plt.plot(x, np.sin(x), 'o')
```

**NOTE : passing `marker = 'o'` will draw the lines and points**

```python
for marker in ['o', '.', ',', '^', '<', '>', 's', 'd', 'p']:
    plt.plot(np.random.rand(5), np.random.rand(5), marker, label="marker='{0}'".format(marker))

plt.xlim(0, 2)
plt.legend(loc='center right')
```

**Other Keyword Arguments of `plt.plot`**

```python
plt.figure(figsize=(10, 5))

plt.plot(x, np.sin(x), 
         color="gray", 
         linestyle="solid", 
         linewidth=5, 
         marker="p", 
         markerfacecolor="white", 
         markeredgecolor="black", 
         markersize=10, 
         markeredgewidth=2)
```

**Note the Shorthand for aforementioned Keyword arguments**

```python
plt.figure(figsize=(10, 5))

plt.plot(x, np.sin(x), 
         c="gray", 
         ls="solid", 
         lw=5, 
         marker="p", 
         mfc="white", 
         mec="black", 
         ms=10, 
         mew=2)
```

## Density and Contour Plots

```python
xmin, xmax, ymin, ymax = 0, 5, 0, 5

x = np.linspace(xmin, xmax, 100)
y = np.linspace(ymin, ymax, 100)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
```

* ### `plt.contour`

```python
plt.contour(X, Y, Z, colors='k')
plt.colorbar()
```

**NOTE:** When single color is used, positive values (peaks) are represented by solid lines whereas negative values (valleys) are represented by dashed lines. We can color-code the lines by specifying a colormap.

```python
plt.contour(X, Y, Z, cmap='RdGy')
plt.colorbar()
```

**Draw more lines in between**

```python
plt.contour(X, Y, Z, levels=15, cmap='RdGy')
plt.colorbar()
```

**NOTE:** The spaces between lines are quite distracting. So we mostly use filled contour plots
    <hr>

* ### `plt.contourf`

```python
plt.contourf(X, Y, Z, levels=15, cmap='RdGy')
plt.colorbar()
```

**NOTE:** The colour steps are still a bit discrete rather than continuous. A better way is to use `plt.imshow`, which interprets 2D grid of data as an image.
<hr>

* ### `plt.imshow`

```python
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy')
plt.colorbar()
```

**NOTE:** 
* No `X` and `Y` grid. Manually need to specify the extent (`[xmin, xmax, ymin, ymax]`) of the image.
* Image is inverted horizontally since default origin (0, 0) is in upper left for a standard image. Manually need to change origin to lower left

```python
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy', origin='lower')
plt.colorbar()
```

**Combine Contour plots and Image plots**

```python
plt.figure(figsize=(7, 7))

contours = plt.contour(X, Y, Z, levels=3, colors='k')   # store the contour lines returned by the plot
plt.clabel(contours, fontsize=8, inline=True)   # customise contour laber
plt.imshow(Z, extent=[xmin, xmax, ymin, ymax], cmap='RdGy', origin='lower', alpha=0.5)
plt.colorbar()
```

## Histograms and Binings

```python
data = np.random.randn(1000)

counts, bin_edges = np.histogram(data)
print("Counts: ", counts)
print("Bin egdes:\n", bin_edges)
```

```python
plt.hist(data)
```

**Other options to customise histogram**

```python
histtypes = ['bar', 'step', 'stepfilled']

plt.figure(figsize=(20, 5))

for i, htype in enumerate(histtypes):
    plt.subplot(1, 3, i + 1)
    plt.hist(data, bins=25, color='c', edgecolor='b', linewidth=3, alpha=0.7, histtype=htype)
```

### Over-plotting multiple histograms

```python
x1 = np.random.normal(0, 0.5, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

colors = ['c', 'y', 'm']
kwargs = dict(bins=25, edgecolor='k', linewidth=2, alpha=0.5, histtype='stepfilled')
```

```python
for color, data in zip(colors, [x1, x2, x3]):
    plt.hist(data, **kwargs)
```

### Two-Dimensional Histogram

* #### **`plt.hist2d`**

```python
mean = [0, 0] 
cov = [[1, 1], [1, 2]]

x, y = np.random.multivariate_normal(mean, cov, 1000).T
```

```python
_ = plt.hist2d(x, y, bins=15, cmap='Blues') # hiding the output array and only displaying the plot
plt.colorbar(label='counts in bins')
```

* #### **`plt.hexbin`**


```python
plt.hexbin(x, y, bins=20, gridsize=15, cmap='Blues')  # gridsize analogous to bins
plt.colorbar(label='counts in bins')
```
