```python
import matplotlib.pyplot as plt
plt.style.use('classic')

import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = np.cumsum(np.random.randn(100, 6), 0)
```

```python
plt.plot(x, y)
plt.legend(labels='ABCDEF', ncol=2, loc='best')
```

**Overwrite Matplotlib's default paramaters with Seaborn**

```python
import seaborn as sns
sns.set() 
```

```python
plt.plot(x, y)
plt.legend(labels='ABCDEF', ncol=2, loc='best')
```

```python
sns.relplot(x = x, y = np.sin(x)) # default = scatter
```

```python
sns.relplot(x = x, y = np.sin(x), kind='line')
```

```python
data = np.random.multivariate_normal([0, 1, 3], [[4, 2, 1], [2, 2, 2], [1, 2, 4]], 1000)
data = pd.DataFrame(data, columns=['x', 'y', 'z'])
```

```python
for col in data.columns:
    plt.hist(data[col], alpha=0.3)
```

```python
for col in 'xyz':
    sns.kdeplot(data[col])
```

```python
for col, c in zip('xyz', 'brg'):
    sns.kdeplot(data[col], shade=True, color=c)
```
